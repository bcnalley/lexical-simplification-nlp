```python
import pandas as pd
import matplotlib.pyplot as plt
import PIL
import seaborn as sns

original_df = pd.read_csv('Data/scores_o.csv')
group1_df = pd.read_csv('Data/scores_1.csv')
group2_df = pd.read_csv('Data/scores_2.csv')
```

## Side-by-Side Box Plots Including Data Points

```python
# Add group column and combine group1 and group2 data frames

one_group = ['One'] * len(group1_df)
two_group = ['Two'] * len(group2_df)

group1_df['Group'] = one_group
group2_df['Group'] = two_group

group1_df = group1_df.drop(columns = 'Unnamed: 0')
group2_df = group2_df.drop(columns = 'Unnamed: 0')

both_groups = pd.concat([group1_df, group2_df])
```

```python
# MSTTR plot

fig, ax = plt.subplots()

sns.set_style('darkgrid')
sns.set_palette('rainbow')
plt.rcParams['figure.dpi'] = 300

sns.boxplot(x = 'Group', y = 'MSTR', data = both_groups, showfliers = False, meanline = True, showmeans = True, ax = ax)
sns.stripplot(x = 'Group', y = 'MSTR', data = both_groups, dodge = True, c = 'black', ax = ax)
ax.axhline(original_df['MSTR'].item(), c = 'r', label = 'Value in Original Story')
ax.set_ylabel('MSTTR')

fig.legend(bbox_to_anchor=(0, 1, 1, 0), loc="lower right", mode="expand", ncol=2)

plt.show()

fig.savefig('Plots/MSTTR_boxplot_points.png', bbox_inches="tight")

# HDD Plot

fig, ax = plt.subplots()

sns.set_style('darkgrid')
sns.set_palette('rainbow')
plt.rcParams['figure.dpi'] = 300

sns.boxplot(x = 'Group', y = 'HDD', data = both_groups, showfliers = False, meanline = True, showmeans = True, ax = ax)
sns.stripplot(x = 'Group', y = 'HDD', data = both_groups, dodge = True, c = 'black', ax = ax)
ax.axhline(original_df['HDD'].item(), c = 'r', label = 'Value in Original Story')
ax.set_ylabel('HDD')

fig.legend(bbox_to_anchor=(0, 1, 1, 0), loc="lower right", mode="expand", ncol=2)

plt.show()

fig.savefig('Plots/HDD_boxplot_points.png', bbox_inches="tight")

# MTLD plot

fig, ax = plt.subplots()

sns.set_style('darkgrid')
sns.set_palette('rainbow')
plt.rcParams['figure.dpi'] = 300

sns.boxplot(x = 'Group', y = 'MTLD bi', data = both_groups, showfliers = False, meanline = True, showmeans = True, ax = ax)
sns.stripplot(x = 'Group', y = 'MTLD bi', data = both_groups, dodge = True, c = 'black', ax = ax)
ax.axhline(original_df['MTLD bi'].item(), c = 'r', label = 'Value in Original Story')
ax.set_ylabel('MTLD')

fig.legend(bbox_to_anchor=(0, 1, 1, 0), loc="lower right", mode="expand", ncol=2)

plt.show()

fig.savefig('Plots/MTLD_boxplot_points.png', bbox_inches="tight")

# Advanced Guiraud plot

fig, ax = plt.subplots()

sns.set_style('darkgrid')
sns.set_palette('rainbow')
plt.rcParams['figure.dpi'] = 300

sns.boxplot(x = 'Group', y = 'Custom AG', data = both_groups, showfliers = False, meanline = True, showmeans = True, ax = ax)
sns.stripplot(x = 'Group', y = 'Custom AG', data = both_groups, dodge = True, c = 'black', ax = ax)
ax.axhline(original_df['Custom AG'].item(), c = 'r', label = 'Value in Original Story')
ax.set_ylabel('Advanced Guiraud')

fig.legend(bbox_to_anchor=(0, 1, 1, 0), loc="lower right", mode="expand", ncol=2)

plt.show()

fig.savefig('Plots/AG_boxplot_points.png', bbox_inches="tight")
```
